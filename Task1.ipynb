```python
#get data ready
import pandas as pd
import numpy as np
data=pd.read_csv("tested.csv")
data.head(5)
```

```python
data.isna().sum()
```

```python
#Fill missing values
data["Cabin"].fillna("missing",inplace=True)
data["Age"].fillna(data["Age"].mean(),inplace=True)
data["Fare"].fillna(data["Fare"].mean(),inplace=True)
```

```python
data.head(5)
```

```python
data.isna().sum()
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
X=data.drop("Survived",axis=1)
y=data["Survived"]
catagorical_feature=["Name","Sex","Ticket","Cabin","Embarked"]
# catagorical_feature=["Name"]
one_hot=OneHotEncoder()
transformer=ColumnTransformer([("one_hot",one_hot,catagorical_feature)],remainder="passthrough")
transmitted_X=transformer.fit_transform(X)
transmitted_X=transmitted_X.toarray()
type(transmitted_X)
```

```python
X=pd.DataFrame(transmitted_X)
X
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
np.random.seed(42)
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
clf=RandomForestClassifier()
clf.fit(x_train,y_train)
```

```python
clf.score(x_test,y_test)
```
